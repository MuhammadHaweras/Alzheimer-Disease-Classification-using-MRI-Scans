```python
import cv2
import os
import glob
import random
import imageio as imgio
import matplotlib.pyplot as plt
```

```python
folder_paths = [
    r'F:\MS CS\1st Semester\Azheimer Classification\combined_images\MildDemented',
    r'F:\MS CS\1st Semester\Azheimer Classification\combined_images\ModerateDemented',
    r'F:\MS CS\1st Semester\Azheimer Classification\combined_images\NonDemented',
    r'F:\MS CS\1st Semester\Azheimer Classification\combined_images\VeryMildDemented',
]
```

```python
output_folder_paths = [
    r'F:\MS CS\1st Semester\Azheimer Classification\new_data_set\MildDemented',
    r'F:\MS CS\1st Semester\Azheimer Classification\new_data_set\ModerateDemented',
    r'F:\MS CS\1st Semester\Azheimer Classification\new_data_set\NonDemented',
    r'F:\MS CS\1st Semester\Azheimer Classification\new_data_set\VeryMildDemented',
]
```

```python
for folder_path, output_folder in zip(folder_paths, output_folder_paths):
    # Get all image file paths in the folder
    image_paths = glob.glob(os.path.join(folder_path, '*'))

    for image_path in image_paths:
        # Read the image in grayscale
        img = cv2.imread(image_path, 0)  # 0 flag for grayscale

        # Resize the image to (224, 224)
        resized_img = cv2.resize(img, (224, 224))

        # Construct the output file path
        file_name = os.path.basename(image_path)
        output_path = os.path.join(output_folder, file_name)

        # Save the processed image
        cv2.imwrite(output_path, resized_img)

print("Image processing completed.")
```

```python

def display_random_images(folder_paths):
    num_folders = len(folder_paths)
    plt.figure(figsize=(12, 8))

    for idx, folder_path in enumerate(folder_paths):
        # Get all image file paths in the folder
        image_paths = glob.glob(os.path.join(folder_path, '*'))

        if image_paths:  # Check if the folder is not empty
            # Select a random image
            random_image_path = random.choice(image_paths)
            img = cv2.imread(random_image_path, cv2.IMREAD_GRAYSCALE)  # Read in grayscale

            # Get the class label from the folder name (last part of the path)
            class_label = os.path.basename(folder_path)

            # Display the image
            plt.subplot(1, num_folders, idx + 1)
            plt.imshow(img, cmap='gray')
            plt.title(class_label)
            plt.axis('off')

        else:
            print(f"No images found in {folder_path}")

    plt.tight_layout()
    plt.show()

display_random_images(folder_paths)
```
